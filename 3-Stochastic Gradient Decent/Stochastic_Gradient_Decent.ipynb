```python
def StochasticGradientDecentSingle(x, y, learningRate = .0001, n_iter = 500):
    lstCost = []
    lstTheta0 = []
    lstTheta1 = []
    lstPredicted_Data = []
    
    theta_0 = 0
    theta_1 = 0
    
    for i in range(n_iter): 
        for j in range(len(x)):
            predicted_data = theta_0 + theta_1 *x[j]
        
            cost = (predicted_data -y[j])**2
            
            grad_theta_0 = predicted_data - y[j]
            grad_theta_1 = (predicted_data - y[j])*x[j]
        
            theta_0 = theta_0 - learningRate * grad_theta_0
            theta_1 = theta_1 - learningRate * grad_theta_1
        
        lstCost.append(cost)
        lstTheta0.append(theta_0)
        lstTheta1.append(theta_1)
        
        lstPredicted_Data.append(theta_0 + theta_1 *x)
        

    return lstTheta0,lstTheta1 ,lstCost,lstPredicted_Data
```

```python
lstTheta0, lstTheta1, lstCost, lstPredictedData = StochasticGradientDecentSingle(x,y)
```

# Plotting cost values over iterations

```python
plt.figure(figsize=(10,7))
ax = plt.axes()
ax.set_facecolor("black")
plt.plot(lstCost, c= "g",  lw = 5)
```
