```python
def MiniBatchGradientDecentSingleNew(x, y, batch_size = 5, learningRate = .0005, n_iter = 300):
    lstCost = []
    lstTheta0 = []
    lstTheta1 = []
    lstPredicted_Data = []
    
    theta_0 = 0
    theta_1 = 0
    
    num_of_batches =int(len(x)/ batch_size)
    
    for i in range(0,n_iter):
        for j in range(0,len(x),num_of_batches):
            predicted_data = theta_0 + theta_1 * x[j:j+num_of_batches]
            #print(predicted_data)
            cost = 1/(2*len(x)) * sum((predicted_data -y[j:j+num_of_batches])**2)
            
            grad_theta_0 = (1/num_of_batches) * sum(predicted_data - y[j:j+num_of_batches])
            grad_theta_1 = (1/num_of_batches) * sum((predicted_data - y[j:j+num_of_batches])*x[j:j+num_of_batches])
            
            theta_0 = theta_0 - learningRate * grad_theta_0
            theta_1 = theta_1 - learningRate * grad_theta_1
            
        lstCost.append(cost)    
        lstTheta0.append(theta_0)
        lstTheta1.append(theta_1)    
        lstPredicted_Data.append(theta_0 + theta_1 *x)
    return lstTheta0,lstTheta1 ,lstCost,lstPredicted_Data
```

```python
lstTheta0, lstTheta1, lstCost, lstPredictedData = MiniBatchGradientDecentSingleNew(x,y)
```

# Plotting cost values over iterations

```python
plt.figure(figsize=(10,7))
ax = plt.axes()
ax.set_facecolor("black")
plt.plot(lstCost, c= "g",  lw = 5)
```

# Plotting the Best fit Line

```python
plt.figure(figsize=(10,7))
ax = plt.axes()
ax.set_facecolor("black")
plt.scatter(x,y,marker= "x",c="g",lw = 3)
plt.plot(x,lstPredictedData[-1],marker= None,c="gray",lw = 3)
```
