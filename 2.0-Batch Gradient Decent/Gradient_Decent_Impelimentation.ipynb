```python
def GradientDecent(x, y, learningRate = .001 ,n_iter = 100):
    lstCost = []
    lstTheta0 = []
    lstTheta1 = []
    lstPredictedData = []
    
    theta_0 = 0
    theta_1 = 0
    for i in range(n_iter):  
        predicted_data = theta_0 + theta_1 *x 
        
        cost = 1/(2*len(x)) * sum((predicted_data -y)**2)
        
        grad_theta_0 = (1/len(x)) * sum(predicted_data - y)
        grad_theta_1 = (1/len(x)) * sum((predicted_data - y)*x)
        
        theta_0 = theta_0 - learningRate * grad_theta_0
        theta_1 = theta_1 - learningRate * grad_theta_1
        lstCost.append(cost)
        lstTheta0.append(theta_0)
        lstTheta1.append(theta_1)
        lstPredictedData.append(predicted_data)
    
    return lstTheta0, lstTheta1, lstCost, lstPredictedData 
```

```python
lstTheta0, lstTheta1, lstCost, lstPredictedData = GradientDecent(x,y)
```

# Plotting cost values over iterations

```python
plt.figure(figsize=(10,7))
ax = plt.axes()
ax.set_facecolor("black")
plt.plot(lstCost, c= "g",  lw = 5)
```

# Plotting the Best fit Line

```python
plt.figure(figsize=(10,7))
ax = plt.axes()
ax.set_facecolor("black")
plt.scatter(x,y,marker= "x",c="g",lw = 3)
plt.plot(x,lstPredictedData[-1],marker= None,c="gray",lw = 3)
```
